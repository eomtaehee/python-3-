````python
# 꺽은선형 그래프 그리기
# plot

# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')

# 그래프 데이터 설정
mon = ['mar','apr','may','jun','jul'] #x축
sales = [1, 5, 7, 3, 7] #y축

# 그래프 제목
plt.title('월별 판매 실적 : 꺽은선형 그래프')
# 그래프 그리기  x축값, y축값, 색상, 범례
plt.plot(mon, sales, color='red', label='단위:개수')
# 범례위치 지정
plt.legend() #default 왼쪽위
plt.show()
````

````python
# 막대그래프 그리기
# bar

# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')

# 그래프 데이터 설정
mon = ['mar','apr','may','jun','jul'] #x축
sales = [1, 5, 7, 3, 7] #y축

# 제목
plt.title('월별 판매 실적 : 세로형 막대 그래프')

# 막대그리프 그리기
plt.bar(mon, sales, color='green', label='단위:개수')
plt.legend()
plt.show()
````

## range(시작숫자, 종료숫자,간격)
시작숫자부터 종료숫자까지 간격을 두어 숫자를 선택함 </br>
```
range(2, 11, 2)
```
2부터 11가지 간격 2를 두고 숫자 선택
-> 2, 4, 6, 8, 10

````python
plt.title('월별 판매 실적 : x축 데이터 범위 설정')
plt.bar(range(2,12,2), sales, color='blue')
plt.show()
````

````python
# plt.barh() : 막대그래프 - 가로형
plt.title('월별 판매 실적 가로형 막대 그래프')
plt.barh(range(3,8), sales, color = 'hotpink')
plt.show()
````

````python
# 히스토그램 : 빈돗수 나타내는 명령어
# plt.hist()
print(sales)
````

````python
plt.title('판매 실적 히스토그램')
plt.hist(sales)
plt.show()
````

## plt.pie() : 원그래프
```
우리학교 100명의 혈액형을 조사한 결과가 다음과 같다.
A형 = 28명
B형 = 20명
O형 = 40명
AB형 = 12명
```

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family='Malgun Gothic')

# data
b_type = [28, 30, 40, 12]

# 원그래프 pie
plt.pie(b_type)
plt.show()
````

## 원그래프 각 항목에 라벨 붙이기
```
plt.pie(리스트, labels = 라벨리스트)
```

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family='Malgun Gothic')

# data
b_type = [28, 30, 40, 12]
b_name = ['A형', 'B형', 'O형', 'AB형']

# 제목
plt.title('혈액형 비율 : 원형 그래프')
# 원형 그래프 autopct='%1.1f%%'
plt.pie(b_type, labels = b_name, autopct='%1.1f%%')
plt.show()
````

## plt.scatter() : 산점도
국어를 잘하는 친구가영어도 잘할까?
```
국어점수와 영어점수의 상관관계 알아보기

kor = [80, 20, 50, 20, 10, 50, 60, 30, 60]
eng = [90, 40, 60, 40, 10, 30, 50, 70, 90]
```

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family='Malgun Gothic')

# data
kor = [80, 20, 50, 20, 10, 50, 60, 30, 60]
eng = [90, 40, 60, 40, 10, 30, 50, 70, 90]
# 제목
plt.title('국어점수와 영어점수와의 상관관계')

# 산점도 그래프 그리기
plt.scatter(kor, eng, c='indigo')
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.show()
````

````python
# 그래프 라이브러리 가져오기
import matplotlib.pyplot as plt
# 랜덤함수 가져오기
import random

# data
height = []
weight = []

for i in range(100) :
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    
# font
plt.rc('font', family='Malgun Gothic')
# title
plt.title('키와 몸무게의 상관관계')
# 산점도 그래프 scatter
plt.scatter(height, weight, c= height, cmap='RdPu')
plt.colorbar(label='키')
# 축제목
plt.xlabel('키')
plt.ylabel('몸무게')
# 그래프를 보여줘
plt.show()
````
