## plt.scatter() : 산점도
> 국어를 잘하는 친구가 영어도 잘할까?
```
국어점수와 영어점수의 상관관계 알아보기

kor =[80, 20, 50, 20, 10, 50, 60, 30, 60]
eng =[90, 40, 60, 40, 10, 30, 50, 70, 90]
```

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family = 'Malgun Gothic')

#data
kor =[80, 20, 50, 20, 10, 50, 60, 30, 60]
eng = [90, 40, 60, 40, 10, 30, 50, 70, 90]

#제목
plt.title('국어점수와 영어점수와의 상관관계')

# 산점도 그래프 그리기
plt.scatter(kor, eng, c='indigo')
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.show()
````
