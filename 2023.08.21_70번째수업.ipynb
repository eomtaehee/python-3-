## plt.pie : 원그래프
```
우리학교 100명의 혈액형을 조사한 결과가 다름과 같다.
A형 = 28명
B형 = 20명
O형 = 40명
AB형 = 12명
```

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family = 'Malgun Gothic')

# data
b_type = [28, 30, 40, 12]

# 원그래프 pie
plt.pie(b_type)
plt.show()
````

## 원그래프 각 항목에 라벨 붙이기
```
plt.pie(리스트, labels = 라벨리스트)
```

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')

# data labels
b_type = [28,30,40,12]
b_name= ['A형','B형','O형','AB형']

#제목
plt.title('혈액형 비율 : 원형 그래프')
#원형 그래프 autopct='%1.f%%
plt.pie(b_type, labels = b_name, autopct='%1.1f%%')
plt.show()
````

### plt.scatter() : 산점도

> 국어를 잘하는 친구가 영어도 잘할까?
```
국어점수와 영어점수의 상관관계 알아보기

kor = [80, 20, 50, 20, 10, 50, 60, 30, 60]
eng = [90, 40, 60, 40, 10, 30, 50, 70, 90]
```

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')

# bata
kor = [80, 20, 50, 20, 10, 50, 60, 30, 60]
eng = [90, 40, 60, 40, 10, 30, 50, 70, 90]


# 산점도 그래프 그리기
plt.scatter(kor, eng)
plt.show()
````
