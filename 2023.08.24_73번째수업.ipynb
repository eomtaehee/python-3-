## 원형그래프 각 항목에 라벨 붙이기
```
plt.pie(리스트, labels = 라벨리스트)
```

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family='Malgun Gothic')

#data Labels
b_type = [28, 30, 40, 12]
b_name = ['A형','B형','O형','AB형']

# 제목
plt.title('혈액형 비율 : 원형 그래프')
# 원형 그래프 autopct='%1.1f%%'
plt.pie(b_type, labels = b_name, autopct='%1.1f%%')
plt.show()
````

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')

# data
kor =[80, 20, 50, 20, 10, 50, 60, 30, 60]
eng =[90, 40, 60, 40, 10, 30, 50, 70, 90]
# 제목
plt.title('국어점수와 영어점수와의 상관관계')

# 산점도 그래프 그리기
plt.scatter(kor, eng, c='indigo')
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.show()
````

````python
# 그래프 라이브러리 가져오기
import matplotlib.pyplot as plt
# 랜덤함수 가져오기
import random

# data
height = []
weight = []

for i in range(100) :
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    
# font
plt.rc('font', family='Malgun Gothic')
# title
plt.title('키와 몸무게의 상관관계')
# 산점도 그래프 scatter
plt.scatter(height, weight, c= height, cmap='RdPu')
plt.colorbar(label='키')
# 축제목
plt.xlabel('키')
plt.ylabel('몸무게')
# 그래프를 보여줘
plt.show()
````
