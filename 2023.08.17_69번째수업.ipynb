### range(시작숫자, 종료숫자, 간격)

시작숫자부터 종료숫자까지 간격을 두어 숫자를 선택함

```
range(2, 11, 2,)
```
2부터 11까지 간격 2를 두고 숫자 선택
-> 2, 4, 6, 8, 10

````python
# plt.barh() : 막대그래프 - 가로형
plt.title('월별 판매 실적 가로형 막대 그래프')
plt.bar(range(2, 11, 2,), slaes, color = 'pink')
plt.show()
````

````python
# plt.barh() : 막대그래프 - 가로형
plt.title('월별 판매 실적 가로형 막대 그래프')
plt.barh(range(3,8), slaes, color = 'hotpink')
plt.show()
````

````python
plt.title('판매 실적 히스토그램')
plt.hist(slaes)
plt.show()
````

### plt.pie() : 원그래프
```
우리학교 100명의 혈액형을 조사한 결과가 다음과 같다.
A형 = 28명
B형 = 20명
O형 = 40명
AB형 = 12명
```

````python
# data
b_type = [28, 30, 40, 12]

# 원그래프 pie
plt.pie(b_type)
plt.show()
````

````python
#bate
b_type = [28,30,40,12]
b_name= ['A형','B형','O형','AB형']

#제목
plt.title('혈액형 비율 : 원형 그래프')
#원형 그래프 autopct='%1.f%%
plt.pie(b_type, labels = b_name, autopct='%1.1f%%')
plt.show()
````
